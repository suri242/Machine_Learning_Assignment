## Analysis IQR value of the columns and making outlier free columnsL of the dataset.

### Creating a function for IQR.

```python
def IQR(col):
    Q3=np.quantile(df[col],0.75)
    Q1=np.quantile(df[col],0.25)
    
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q1 + 1.5*IQR
    
    suri = df[((df[col] >= lower_limit) & (df[col] <=upper_limit ))]
    
    df[col] = np.where(df[col] > upper_limit,
        upper_limit,
        np.where(df[col] < lower_limit,
        lower_limit,
        df[col]))
    
    plt.figure(figsize=(4,4))
    plt.boxplot(df[col])
    plt.xlabel(f"{col}")
    plt.title(f"Outlier_free_{col}")
    plt.show()
```

## Separating the X and the Y values

```python
X= df.drop(["target"], axis=1)
y= df.target

print(X.shape)
print(y.shape)
```

## Training and Testing Data Bifurcation using Scikit-Learn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=4)
```

## Building a function for evaluation scores.

```python
def evaluate_score(ss):
    from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
    import math
    from sklearn.metrics import f1_score, matthews_corrcoef , classification_report
    ss.fit(x_train,y_train)
    y_pred=ss.predict(x_test)
    
    print(f"*---------------------------------------Score_{ss}-------------------------------------*")
    rr= accuracy_score(y_test, y_pred)
    print(f"The Accuracy_Score is :- {rr} \n")

    prec = precision_score(y_test, y_pred)
    print(f"The Precision_Score  is :- {prec}\n")

    rec = recall_score(y_test, y_pred)
    print(f"The Recall_Score is :- {rec}\n")

    f1 = f1_score(y_test, y_pred)
    print(f"The F1-Score is :- {f1}\n")

    MCC = matthews_corrcoef(y_test, y_pred)
    print(f"The Matthews correlation coefficient is :- {MCC}\n")

    cr=classification_report(y_test, y_pred)
    print(f"The Classification_Report is :- \n {cr}\n")
  
    print(f"*------------------------***Confusion_Matrix for_{ss}***------------------------------* \n")
    cm=confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,4))
    sns.heatmap(cm,annot=True, xticklabels=["No_Spill","Spill"], yticklabels=["No_Spill","Spill"])
    plt.title("Confusion matrix")
    plt.ylabel('True_class')
    plt.xlabel('Predicted_class')
    plt.show()
```
